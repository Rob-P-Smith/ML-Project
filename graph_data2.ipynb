# Data Loading and Initial Setup
Import pandas, matplotlib, and load the CSV file. Create a year range array for x-axis.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = './Outputs/pop_sea_temp_fleet_amazon_oil.csv'
data = pd.read_csv(file_path)

# Create a year range array for x-axis
years = list(range(1880, 2022))

# Display the first few rows of the dataframe to understand its structure
data.head()
```

# Create Individual Line Plots
Iterate through each column in the dataset and create line plots using matplotlib.

```python
# Create Individual Line Plots
for column in data.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(years, data[column], label=column)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(column.replace('_', ' ').title())
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Format Plots with Labels and Legends
Add proper labels, titles, and legends to each plot. Convert column names to plain English for titles.

```python
# Format Plots with Labels and Legends
for column in data.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(years, data[column], label=column)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(column.replace('_', ' ').title())
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Multiple Subplots Layout
Create a figure with multiple subplots to display all graphs in a grid layout.

```python
# Multiple Subplots Layout
import math

# Determine the number of columns and rows for the subplots grid
num_columns = 3
num_rows = math.ceil(len(data.columns) / num_columns)

# Create a figure with multiple subplots
fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, 15))
axes = axes.flatten()

# Plot each column in a separate subplot
for i, column in enumerate(data.columns):
    axes[i].plot(years, data[column], label=column)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('Value')
    axes[i].set_title(column.replace('_', ' ').title())
    axes[i].legend()
    axes[i].grid(True)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

# Save Generated Plots
Save the generated plots as image files for future reference.

```python
# Save Generated Plots

import os

# Create a directory to save the plots if it doesn't exist
output_dir = './Generated_Plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Save each plot as an image file
for column in data.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(years, data[column], label=column)
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(column.replace('_', ' ').title())
    plt.legend()
    plt.grid(True)
    plt.savefig(f'{output_dir}/{column}.png')
    plt.close()
```
